```python
#Selecting best Features using Select K Best
X = df.drop(['subclass',],axis=1)
y=df['subclass']
```

```python
#i want to take top k best columns in the data frame using selectKBest k = 10
from sklearn.feature_selection import SelectKBest, f_classif

# Keep only numeric columns
X_numeric = X.select_dtypes(include=['int64', 'float64'])

X_numeric = X_numeric.fillna(X_numeric.mean())

#assuming Xand y are your data and target variables
selector = SelectKBest(score_func=f_classif, k=10)  #select top 10 features

#fit selctor to data
X_selected = selector.fit_transform(X_numeric, y)

#get the name of selected features
selected_features = X_numeric.columns[selector.get_support()]

#print the selected features
print("Selected Features:", list(selected_features))
```

```python
#Assuming your target column is 'subclass' in your DataFrame 'df'
X=df.drop(['subclass','class'], axis=1)
y= df['subclass']

# Convert categorical columns to numeric
X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded = X_encoded.fillna(X_encoded.mean())

#Initialize SMOTE
smote = SMOTE(random_state=42)

#Perform SMOTE oversampling
X_resampled, y_resampled = smote.fit_resample(X_encoded,y)

#Check the new value counts
print(pd.Series(y_resampled).value_counts())
```

```python
# Select required columns
df1 = df[[
    'u', 'z', 'modelFlux_z',
    'petroRad_g', 'petroRad_r', 'petroFlux_z',
    'petroR50_u', 'petroR50_g', 'petroR50_i', 'petroR50_r',
    'subclass'
]]

# Split features and target
X = df1.drop('subclass', axis=1)
y = df1['subclass']

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Check shapes
print(x_train.shape, x_test.shape)
print(y_train.value_counts())
```

```python
#Decision Tree Classifier :
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Initialize the classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier
clf.fit(x_train, y_train)

# Make predictions on test data
y_pred = clf.predict(x_test)

# Evaluate the classifier
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

#Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score
)

# Initialize Logistic Regression
log = LogisticRegression(max_iter=1000, random_state=42)

# Train the model
log.fit(x_train, y_train)

# Predict on test data
y_pred = log.predict(x_test)

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("--------------------------------------------------")

# Classification Report
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

# Evaluation Metrics
print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred, average='weighted'))
print("Recall   :", recall_score(y_test, y_pred, average='weighted'))
print("F1 Score :", f1_score(y_test, y_pred, average='weighted'))
```

#Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score
)

# Train the  Random Forest classifier
RF = RandomForestClassifier()

# Train model
RF.fit(x_train, y_train)

# Predictions
RFtrain = RF.predict(x_train)
RFtest = RF.predict(x_test)

# printing classification report, confusion matrix
print("TRAIN PERFORMANCE")
print(confusion_matrix(RFtrain,y_train))
print(classification_report(RFtrain, y_train))

print("\nTEST PERFORMANCE")
print(confusion_matrix(RFtest,y_test))
print(classification_report(RFtest,y_test))
```

```python
print("Accuracy:", accuracy_score(RFtrain,y_train))
print("Accuracy:", accuracy_score(RFtest,y_test))
```
